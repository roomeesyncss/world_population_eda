```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path='world_population.csv'
world_population_df=pd.read_csv(file_path)
world_population_df
```

```python
# Calculate the percentage change in population for each country from 1980 to 2022
world_population_df['Population Change (%)'] = ((world_population_df['2022 Population'] - world_population_df['1980 Population']) / world_population_df['1980 Population']) * 100

#sorting values accordingly 
world_population_df_sorted = world_population_df.sort_values('Population Change (%)', ascending=False)
world_population_df_sorted

```

```python
plt.figure(figsize=(102, 80))
barplot = sns.barplot(x='Population Change (%)', y='Country/Territory', data=world_population_df_sorted.head(20), palette='viridis')

# Adjusting the font size of x and y-axis labels
barplot.set_xlabel('Percentage Change in Population', fontsize=80)
barplot.set_ylabel('Country', fontsize=80)

# Adjusting the font size of x and y-axis tick labels
plt.xticks(fontsize=70)
plt.yticks(fontsize=70)

plt.title('Top 20 Countries by Percentage Change in Population (1980-2022)', fontsize=100)
plt.show()

```

```python

world_population_df_sorted_by_growth = world_population_df.sort_values('Growth Rate', ascending=False)

# Select the top 10 countries
top_10_growth_countries = world_population_df_sorted_by_growth.head(10)

# Display the top 10 countries with the highest population growth rate
print(top_10_growth_countries[['Country/Territory', 'Growth Rate']])

plt.figure(figsize=(102, 80))
barplot = sns.barplot(x='Growth Rate', y='Country/Territory', data=top_10_growth_countries, palette='coolwarm')

# Adjusting the font size of x and y-axis labels
barplot.set_xlabel('Growth Rate', fontsize=74)
barplot.set_ylabel('Country', fontsize=74)

# Adjusting the font size of x and y-axis tick labels
plt.xticks(fontsize=72)
plt.yticks(fontsize=72)

plt.title('Top 10 Countries by Population Growth Rate', fontsize=96)
plt.show()
```

```python
import geopandas as gpd
import contextily as ctx

# Load the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world map with the population data
world_population_gdf = world.merge(world_population_df, how='left', left_on='name', right_on='Country/Territory')

# Plot the map with population density
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world_population_gdf.plot(column='Density (per km\u00b2)', ax=ax, legend=True,
                         legend_kwds={'label': "Population Density (per km\u00b2)",
                                      'orientation': "horizontal"},
                         cmap='OrRd')
ctx.add_basemap(ax, crs=world.crs.to_string(), source=ctx.providers.CartoDB.Positron)
ax.set_axis_off()
```

```python
# Correcting the code to ensure the legend is created and setting the font size and weight
import geopandas as gpd
import contextily as ctx

# Load the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world map with the population data
world_population_gdf = world.merge(world_population_df, how='left', left_on='name', right_on='Country/Territory')

# Plot the map with population density
fig, ax = plt.subplots(1, 1, figsize=(20, 12))
world_population_gdf.plot(column='Density (per km\u00b2)', ax=ax, legend=True,
                         legend_kwds={'label': "Population Density (per km\u00b2)",
                                      'orientation': "horizontal"},
                         cmap='OrRd')
ctx.add_basemap(ax, crs=world.crs.to_string(), source=ctx.providers.CartoDB.Positron)

# Check if the legend is created
if ax.get_legend() is not None:
    leg = ax.get_legend()
    leg.set_title('Population Density (per km\u00b2)', prop={'size': 30, 'weight': 'bold'})
    for label in leg.get_texts():
        label.set_fontsize(30)
        label.set_fontweight('bold')

# Enhancing font size for better readability
ax.set_title('World Population Distribution', fontsize=20, fontweight='bold')
ax.set_xlabel('Longitude', fontsize=20, fontweight='bold')
ax.set_ylabel('Latitude', fontsize=20, fontweight='bold')

ax.set_axis_off()
plt.show()
```

```python
# Calculate the countries with the highest and lowest population densities
# Sort the dataframe by the density
world_population_df_sorted_by_density = world_population_df.sort_values('Density (per km\u00b2)', ascending=False)

# Select the country with the highest density
top_density_country = world_population_df_sorted_by_density.head(1)
# Select the country with the lowest density
lowest_density_country = world_population_df_sorted_by_density.tail(1)

# Display the countries with the highest and lowest population densities
print('Country with the highest population density:\n', top_density_country[['Country/Territory', 'Density (per km\u00b2)']])
print('\nCountry with the lowest population density:\n', lowest_density_country[['Country/Territory', 'Density (per km\u00b2)']])
```

```python
# Create a bar plot for the countries with the highest and lowest population densities
# We will include a few more countries for context
additional_countries = 5
world_population_df_sorted_by_density_head = world_population_df_sorted_by_density.head(additional_countries)
world_population_df_sorted_by_density_tail = world_population_df_sorted_by_density.tail(additional_countries)

# Combine the dataframes
extreme_density_countries = pd.concat([world_population_df_sorted_by_density_head, world_population_df_sorted_by_density_tail])

# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(x='Density (per km\u00b2)', y='Country/Territory', data=extreme_density_countries, palette='Spectral')
plt.title('Countries with the Highest and Lowest Population Densities')
plt.xlabel('Density (per km\u00b2)')
plt.ylabel('Country')
plt.xscale('log') # Use a log scale for better visualization
plt.show()
```

```python
# Population growth trajectory for selected countries
selected_countries = ['China', 'Syria', 'United States', 'Brazil', 'Nigeria']
plt.figure(figsize=(14, 8))

years = world_population_df.columns[5:]  # Assuming years start from the 6th column

for country in selected_countries:
    country_population = world_population_df.loc[world_population_df['Country/Territory'] == country, '1970 Population':].values.flatten()
    plt.plot(years[:len(country_population)], country_population, label=country)

plt.title('Population Growth Trajectory', fontweight='bold', fontsize=16)
plt.xlabel('Year', fontweight='bold', fontsize=14)
plt.ylabel('Population', fontweight='bold', fontsize=14)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Population pyramid for selected years
selected_years = ['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '2020 Population', '2022 Population']
plt.figure(figsize=(14, 8))
sns.histplot(data=world_population_df[selected_years], kde=True, multiple='stack', palette='viridis', binwidth=50000000)
plt.title('Population Pyramid Over Time', fontweight='bold', fontsize=16)
plt.xlabel('Population', fontweight='bold', fontsize=14)
plt.ylabel('Frequency', fontweight='bold', fontsize=14)
plt.show()
```
